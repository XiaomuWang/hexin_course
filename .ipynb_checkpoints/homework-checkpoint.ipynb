## 任务1 多种滤波操作
图像滤波可以更改或者增强图像，通过滤波，可以强调一些特征或者去除图像中一些不需要的部分。  
滤波是一个邻域操作算子，利用给定周围的像素的值决定此像素的最终输出值。  
本文使用了几个常用的滤波操作供大家学习交流。  
有同学提出边界点如何计算有以下三种方法处理：1.最外面补充0；2.直接放弃无法计算的边缘点；3.把已有的点拷贝到另一面的对应位置，这种方法效果最好。

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("week1_homework.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original image")
plt.subplot(1,2,2)
plt.imshow(gray_img,cmap="gray")
plt.title("Gray image")
```

### 1.均值滤波（normalized box filter）
用其像素点周围像素的平均值代替原中心像素值，在滤除噪声的同时也会滤掉图像的边缘信息。  
opencv函数：blur()或者boxfilter().

```python
kernel_size = 23
blur = cv2.blur(img,(kernel_size,kernel_size))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original image")
plt.subplot(1,2,2)
plt.imshow(blur)
plt.title("Blurred image")
```

### 2.高斯滤波（gaussian filter）
根据二维的正态分布的密度函数即高斯函数，可以计算得出每个点的权重，然后将周围像素值乘以权重求和，再除以权重之和，得到中心点像素值。  
如果原图是彩色图片，需要对三个通道分别做高斯模糊。  
最常用的滤波器，具有可分离性质，可以把二维高斯运算转换为一维高斯运算，为低通滤波器，有效去除高斯噪声.  
opencv函数：Gaussianblur().  
我们应该指定X和Y方向的标准偏差，分别为sigmaX和sigmaY，如果仅指定sigmaX，则sigmaX和sigmaY相同，如果两个都为0则根据内核大小进行计算。

```python
kernel_size = 23
guassian_blur = cv2.GaussianBlur(img,(kernel_size,kernel_size),0)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original image")
plt.subplot(1,2,2)
plt.imshow(guassian_blur)
plt.title("Gaussian Blurred image")
```

### 3.中值滤波（median filter）
用像素周围邻域像素集中的中值代替原像素值。中值滤波在去除椒盐噪声和斑块噪声时，效果十分明显。  
opencv函数：medianBlur().

```python
kernel_size = 23
median_blur = cv2.medianBlur(img,kernel_size)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original image")
plt.subplot(1,2,2)
plt.imshow(guassian_blur)
plt.title("Median Blurred image")
```

### 4.双边滤波（bilateral filter）
原理：一个与空间距离相关的高斯函数与一个灰度距离相关的高斯函数相乘，两者权重系数相乘，得到最终的卷积模板。  
先根据像素值对要用来进行滤波的邻域做一个分割或者分类，再给该点所属类别的像素点赋予较高的权重，然后进行邻域加权求和，得到最终结果。  
判断所属类别是加入了灰度信息的权重距离矩阵，即在邻域内接近中心点灰度值的点的权值更大，灰度差相差大的点的权重越小。  
可以理解为仅考虑与中心点像素相似强度的那些像素进行模糊处理，由于边缘的像素强度变化较大，因此可以保留边缘信息。
参数：空间域σ的选择，其值越大，边缘越模糊，其值越小，边缘越清晰。  
与高斯滤波相比在平滑图像时能够很好的保留边缘特性，因为需要灰度信息确定系数。所以运算速度较慢。

```python
kernel_size = 33
#彩色空间,其值越大滤镜效果越强
sigma_color =100
#坐标空间
sigma_space = 100
bilateral_filter = cv2.bilateralFilter(img,kernel_size,sigma_color,sigma_space)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original image")
plt.subplot(1,2,2)
plt.imshow(bilateral_filter)
plt.title("Bilateral Filter image")
```

### 滤波算法解决思路
通过多种滤波器和各种图像运算结合的思想，生成最终的图像

```python
import numpy as np
def beauty_img(img):
    #磨皮算法公式：
    #Dest =(Src * (100 - Opacity) + (Src + 2 * GuassBlur(EPFFilter(Src) - Src + 128) - 256) * Opacity) /100 ;
    #具体可以参考文章：https://my.oschina.net/wujux/blog/1563461
    #参考了上面的文章，并进行了优化
    
    #双边滤波中每个像素邻域的直径，在这里方便统一学习，设置为kernel_size
    kernel_size = 15 
    sigma_color =37.5
    sigma_space = 37.5
    p =0.1
    bilateral_filter = cv2.bilateralFilter(img,kernel_size,sigma_color,sigma_space)
    #计算两个矩阵之间的差
    subtract = cv2.subtract(bilateral_filter,img)
#     add1 = cv2.add(subtract,(10,10,10,128))
#     guassian_blur = cv2.GaussianBlur(add1,(1,1),0)
    guassian_blur = cv2.GaussianBlur(subtract,(1,1),0)
    
    #输出为形状和img一致的矩阵，其元素全部为0
    zero1 = np.zeros_like(img)    
    #图像叠加
    add2 = cv2.add(img,guassian_blur)
    

    dst = np.zeros_like(img)
    #图像融合，可以设置图片的透明度
    dst = cv2.addWeighted(img,p,add2,1-p,0.0)
#     dst = cv2.add(dst,(10, 10, 10,255))
    return dst


plt.subplot(1,2,1)
beauty_img1 = beauty_img(img)
plt.imshow(beauty_img1)
plt.title("Beauty of original image")
plt.subplot(1,2,2)
beauty_img2 = beauty_img(bilateral_filter)
plt.imshow(beauty_img2)
plt.title("Beauty of bilateral filter image")
```

### 源码实现hog特征，供大家参考学习
https://blog.csdn.net/ppp8300885/article/details/71078555

方向梯度直方图（Histogram of Oriented Gradient, HOG）特征是一种在计算机视觉和图像处理中用来进行物体检测的描述子。  
实现过程：简单来说，首先需要将图像分成小的连通区域，称之为细胞单元。然后采集细胞单元中各像素点的梯度或边缘的方向直方图。最后把这些直方图组合起来就可以构成特征描述器。

#### HOG特征提取算法的整个实现过程大致如下：

1、读入所需要的检测目标即输入的image  
2、将图像进行灰度化（将输入的彩色的图像的r,g,b值通过特定公式转换为灰度值）  
3、采用Gamma校正法对输入图像进行颜色空间的标准化（归一化）  
4、计算图像每个像素的梯度（包括大小和方向），捕获轮廓信息  
5、统计每个cell的梯度直方图（不同梯度的个数），形成每个cell的descriptor  
6、将每几个cell组成一个block（以3*3为例），一个block内所有cell的特征串联起来得到该block的HOG特征descriptor  
7、将图像image内所有block的HOG特征descriptor串联起来得到该image（检测目标）的HOG特征descriptor，这就是最终分类的特征向量

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt


class Hog_descriptor():
    def __init__(self, img, cell_size=16, bin_size=8):
        self.img = img
        #归一化
        self.img = np.sqrt(img / np.max(img))
        self.img = img * 255
        self.cell_size = cell_size
        self.bin_size = bin_size
        self.angle_unit = 360 / self.bin_size
#         assert type(self.bin_size) == int, "bin_size should be integer,"
#         assert type(self.cell_size) == int, "cell_size should be integer,"
#         assert type(self.angle_unit) == int, "bin_size should be divisible by 360"

    def extract(self):
        height, width = self.img.shape
        gradient_magnitude, gradient_angle = self.global_gradient()
        gradient_magnitude = abs(gradient_magnitude)
        cell_gradient_vector = np.zeros((int(height / self.cell_size),int(width / self.cell_size), self.bin_size))
        for i in range(cell_gradient_vector.shape[0]):
            for j in range(cell_gradient_vector.shape[1]):
                cell_magnitude = gradient_magnitude[i * self.cell_size:(i + 1) * self.cell_size,
                                 j * self.cell_size:(j + 1) * self.cell_size]
                cell_angle = gradient_angle[i * self.cell_size:(i + 1) * self.cell_size,
                             j * self.cell_size:(j + 1) * self.cell_size]
                cell_gradient_vector[i][j] = self.cell_gradient(cell_magnitude, cell_angle)

        hog_image = self.render_gradient(np.zeros([height, width]), cell_gradient_vector)
        hog_vector = []
        #统计bolck的梯度信息
        for i in range(cell_gradient_vector.shape[0] - 1):
            for j in range(cell_gradient_vector.shape[1] - 1):
                block_vector = []
                block_vector.extend(cell_gradient_vector[i][j])
                block_vector.extend(cell_gradient_vector[i][j + 1])
                block_vector.extend(cell_gradient_vector[i + 1][j])
                block_vector.extend(cell_gradient_vector[i + 1][j + 1])
                mag = lambda vector: math.sqrt(sum(i ** 2 for i in vector))
                magnitude = mag(block_vector)
                if magnitude != 0:
                    normalize = lambda block_vector, magnitude: [element / magnitude for element in block_vector]
                    block_vector = normalize(block_vector, magnitude)
                hog_vector.append(block_vector)
        return hog_vector, hog_image
    
    #计算每个像素的梯度
    def global_gradient(self):
        gradient_values_x = cv2.Sobel(self.img, cv2.CV_64F, 1, 0, ksize=5)
        gradient_values_y = cv2.Sobel(self.img, cv2.CV_64F, 0, 1, ksize=5)
        gradient_magnitude = cv2.addWeighted(gradient_values_x, 0.5, gradient_values_y, 0.5, 0)
        gradient_angle = cv2.phase(gradient_values_x, gradient_values_y, angleInDegrees=True)
        return gradient_magnitude, gradient_angle
    
    #为每个细胞单元构建梯度方向直方图
    def cell_gradient(self, cell_magnitude, cell_angle):
        orientation_centers = [0] * self.bin_size
        for i in range(cell_magnitude.shape[0]):
            for j in range(cell_magnitude.shape[1]):
                gradient_strength = cell_magnitude[i][j]
                gradient_angle = cell_angle[i][j]
                min_angle, max_angle, mod = self.get_closest_bins(gradient_angle)
                orientation_centers[min_angle] += (gradient_strength * (1 - (mod / self.angle_unit)))
                orientation_centers[max_angle] += (gradient_strength * (mod / self.angle_unit))
        return orientation_centers

    def get_closest_bins(self, gradient_angle):
        idx = int(gradient_angle / self.angle_unit)
        mod = gradient_angle % self.angle_unit
        return idx, (idx + 1) % self.bin_size, mod
    
    #可视化Cell梯度直方图
    def render_gradient(self, image, cell_gradient):
        cell_width = self.cell_size / 2
        max_mag = np.array(cell_gradient).max()
        for x in range(cell_gradient.shape[0]):
            for y in range(cell_gradient.shape[1]):
                cell_grad = cell_gradient[x][y]
                cell_grad /= max_mag
                angle = 0
                angle_gap = self.angle_unit
                for magnitude in cell_grad:
                    angle_radian = math.radians(angle)
                    x1 = int(x * self.cell_size + magnitude * cell_width * math.cos(angle_radian))
                    y1 = int(y * self.cell_size + magnitude * cell_width * math.sin(angle_radian))
                    x2 = int(x * self.cell_size - magnitude * cell_width * math.cos(angle_radian))
                    y2 = int(y * self.cell_size - magnitude * cell_width * math.sin(angle_radian))
                    cv2.line(image, (y1, x1), (y2, x2), int(255 * math.sqrt(magnitude)))
                    angle += angle_gap
        return image
```

```python
#读入彩色图像，并转化为灰度图像
img = cv2.imread('week1_homework.png', cv2.IMREAD_GRAYSCALE)
hog = Hog_descriptor(img, cell_size=4, bin_size=4)
vector, hog_img = hog.extract()
size = img.shape
print(size)
print(np.array(vector).shape)
plt.figure(figsize=(16,16))
# plt.imshow(hog_img,cmap="gray")
# plt.show()

plt.imshow(img)
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.title("Original image")
plt.subplot(1,2,2)
plt.imshow(hog_img,cmap="gray")
plt.title("Hog image")
```
